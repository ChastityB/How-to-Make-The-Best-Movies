```python
#standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
```

```python
#view 2000 data
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000
```

```python
#view 2001 data
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2001
```

```python
#the first row of both dataframes are 0, let us drop them
df_2000_dropped =  df_2000.iloc[1: , :]
df_2001_dropped = df_2001.iloc[1: , :]
display(df_2000_dropped.head(5), df_2000_dropped.head(5))
```

```python
#merge both dataframes into one
df_merged = pd.concat([df_2000_dropped, df_2001_dropped], ignore_index=True)
display(df_merged.head(5), df_merged.head(5))
```

```python
#only extract the needed information
df = df_merged[['imdb_id', 'budget', 'revenue', 'certification']]
df
```

```python
#used to filter out with boolean financial data
filter_index = (df['budget']>0) | (df['revenue']>0)
```

```python
#just movies that had financial data
plot_df = df[filter_index]
plot_df
```

```python
#sns.set_context(context = 'talk')
```

```python
ax = sns.countplot(x = filter_index);
plt.xlabel('Amount of Movies')
plt.ylabel('Movies')
ax.set_title('Movies With Financial Data', fontsize=20)

#to show the amount of movies
for bar in ax.patches:
    ax.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
```
